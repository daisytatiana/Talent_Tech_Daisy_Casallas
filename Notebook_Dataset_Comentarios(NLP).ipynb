```python
import pandas as pd
from textblob import TextBlob
from googletrans import Translator
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
import pygwalker as pyg
```

```python
#Traduccion español-ingles
def translatecomment(strComment:str):
  translator=Translator()
  strEngComment=translator.translate(strComment, src="es",dest="en")
  return strEngComment.text
```

```python
#obtener Polaridad
def getPolarity(strComment:str):
  strComment=translatecomment(strComment)
  analysis = TextBlob(strComment)
  return analysis.sentiment.polarity
```

```python
  #cargar dataset
df_original=pd.read_csv("/content/Comentarios.csv")
df_original['polaridad']=df_original['Comentario'].apply(getPolarity)
```

```python
df_original.to_csv('resultados.csv', index=False)
```

```python
#Creacion dataframe polaridad resultados positivos y negativos
df_resultados=pd.read_csv("/content/resultados.csv")
df_polaridad_positivos=df_resultados['polaridad']>0
df_polaridad_negativos=df_resultados['polaridad']<0
df_cantidad_positivos=df_polaridad_positivos.count()
df_cantidad_negativos=df_polaridad_negativos.sum()
```

```python
#Grafica del archivo resultados
#https://www.datacamp.com/es/tutorial/seaborn-python-tutorial
data1=pd.DataFrame({
    'Polaridad':['Positivo','Negativo'],
    'Cantidad':[df_cantidad_positivos,df_cantidad_negativos]
})
# Crear el gráfico de violín
plt.figure(figsize=(10, 6))
sns.barplot(x='Polaridad', y='Cantidad', data=data1, palette='viridis')
plt.title('Cantidad de comentarios positivos y negativos')
plt.show()
```
